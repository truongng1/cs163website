```python
import pandas as pd

df = pd.read_csv('https://storage.googleapis.com/cs163-seniorproject.appspot.com/dataset/unemployment_rate/raw_dataset/san_jose_unemployment.txt')

numeric_cols = ['labor force', 'employment', 'unemployment', 'unemployment rate']

for col in numeric_cols:

    df[col] = df[col].astype(str).str.replace(r'\(M\)', '', regex=True).str.strip()
    # Convert to numeric
    if col == 'unemployment rate':
        df[col] = pd.to_numeric(df[col], errors='coerce', downcast='float')
    else:
        df[col] = pd.to_numeric(df[col], errors='coerce', downcast='integer')

print("Missing values:\n", df.isnull().sum())

print("\nData types:\n", df.dtypes)

invalid_rates = df[df['unemployment rate'] < 0]
if not invalid_rates.empty:
    print(f"Warning: {len(invalid_rates)} rows with negative unemployment rates")

df['calculated_labor_force'] = df['employment'] + df['unemployment']
mismatch = df[df['labor force'] != df['calculated_labor_force']]
if not mismatch.empty:
    print(f"Warning: {len(mismatch)} rows where labor force != employment + unemployment")

df = df.drop(columns=['calculated_labor_force'])
```
